# Running Forecast command to assess if it works correctly

```python
import subprocess
import sys
```

```python
forecast_data = [[sys.executable,
        "-m", "MASTER.MASTER_forecast_fields",
        "--init-date", "2026-04-01",
        "--lead-months", "6",
        "--generate-config",
        "--active-months", "6", "7", "8", "9", "10", "11"]]

forecast_sample = [[sys.executable,
                    "-m", "MASTER.MASTER_storm_parallel", 
                    "--forecast", "forecasts_configs/config_m${m}.json",
                    "--basins", "NA",
                    "--years", "1000",
                    "--loop", "10",
                    "--workers", "30",
                    ] for m in range(51)]


commands = [
    [sys.executable, "-m", "MASTER.MASTER_climatology"],
    [sys.executable, "-m", "MASTER.MASTER_preprocessing"],
] + forecast_data + forecast_sample
```

```python

for cmd in commands:
    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=True)
```
